To create a 95% bootstrapped confidence interval directly from the bootstrapped sample means, without using their standard deviation to estimate the standard error of the mean, you can use the percentile method. Here's how it works:

Resample the data with replacement many times (e.g., 1,000 or more) to generate a large number of bootstrapped sample means.

Sort the bootstrapped sample means in ascending order.

Determine the percentiles: For a 95% confidence interval, you want to find the 2.5th percentile and the 97.5th percentile of the sorted bootstrapped means. These percentiles represent the lower and upper bounds of the confidence interval, respectively.

Report the interval: The 95% bootstrapped confidence interval is the range between these two percentile values, which will contain 95% of the bootstrapped means.

In summary:

Resample the data many times to generate bootstrapped means.
Sort those means and extract the 2.5th and 97.5th percentiles to form the confidence interval.
This method directly uses the empirical distribution of the bootstrapped means and does not rely on standard error estimates.

````python
import numpy as np

def bootstrap_confidence_interval(data, num_bootstrap=1000, ci_percent=95, statistic_func=np.mean):
    """
    Generates a bootstrap confidence interval for a given statistic (e.g., mean, median)
    
    Parameters:
    - data: The sample data (numpy array or list)
    - num_bootstrap: Number of bootstrap samples to generate
    - ci_percent: The confidence level (default is 95%)
    - statistic_func: The function to compute the statistic (default is np.mean for mean)
    
    Returns:
    - The confidence interval as a tuple (lower bound, upper bound)
    """
    bootstrap_samples = []

    # Create bootstrap samples and compute the statistic for each
    for _ in range(num_bootstrap):
        bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
        stat = statistic_func(bootstrap_sample)  # Compute the statistic (e.g., mean, median) for this sample
        bootstrap_samples.append(stat)

    # Compute percentiles for the confidence interval
    lower_percentile = (100 - ci_percent) / 2
    upper_percentile = 100 - lower_percentile
    lower_bound = np.percentile(bootstrap_samples, lower_percentile)
    upper_bound = np.percentile(bootstrap_samples, upper_percentile)

    return lower_bound, upper_bound

# Example data (sample)
data = [5, 7, 8, 12, 14, 15, 18, 21, 25, 30]

# Generate a 95% bootstrap confidence interval for the population mean
mean_ci = bootstrap_confidence_interval(data, num_bootstrap=1000, ci_percent=95, statistic_func=np.mean)
print(f"95% Bootstrap Confidence Interval for the Mean: {mean_ci}")

# To generate a 95% bootstrap confidence interval for the population median instead:
median_ci = bootstrap_confidence_interval(data, num_bootstrap=1000, ci_percent=95, statistic_func=np.median)
print(f"95% Bootstrap Confidence Interval for the Median: {median_ci}")
````

bootstrap_confidence_interval Function:

data: Your sample data.
num_bootstrap: The number of bootstrap resamples (set to 1,000 by default).
ci_percent: The confidence interval level (default is 95%).
statistic_func: The function used to compute the statistic. By default, it computes the mean (np.mean), but you can pass any other statistic (e.g., np.median).
Main Logic:

A loop generates bootstrap samples by randomly resampling from the original data with replacement.
For each sample, it computes the statistic (mean, median, etc.) and stores the result.
Once all bootstrap samples are collected, the percentiles for the desired confidence level (95%) are computed to determine the lower and upper bounds of the confidence interval.
Changing the Statistic:

To compute a confidence interval for the median or any other statistic, simply change the statistic_func argument. For the median, use np.median.

Population Parameter: This is a fixed value that describes some characteristic of the entire population (e.g., the true mean or proportion). It's usually unknown because it's impractical to collect data from the entire population.

Estimation of Unknown Population Parameters:

1. Start with a Sample Data Set
You begin with a single sample from the population, often referred to as your observed data. This sample contains n observations, and it is assumed to be representative of the population.
2. Resample with Replacement
From this original sample, you randomly draw new samples of size n (same size as the original sample), with replacement. This means that each time you pick a data point, you return it to the dataset so it can be chosen again. This creates what's called a bootstrap sample.
Repeat this process multiple times (e.g., 1,000 or 10,000 times) to generate many bootstrap samples. Each sample is different because of the random resampling.
3. Calculate the Statistic of Interest
For each bootstrap sample, calculate the statistic of interest (e.g., mean, median, standard deviation, or regression coefficient). This results in a distribution of the statistic based on the resampled data.
4. Create a Distribution of the Statistic
After computing the statistic for each bootstrap sample, you will have a large number of estimates of the statistic. These estimates form the bootstrap distribution of the statistic.
5. Estimate Confidence Intervals or Bias
Using the bootstrap distribution, you can estimate:
Confidence intervals: For example, to create a 95% confidence interval, you can take the 2.5th percentile and the 97.5th percentile of the bootstrap distribution.
Bias: You can assess how much the sample statistic is biased relative to the true population parameter by comparing the bootstrap estimate to the original sample estimate.
6. Interpret the Results
The resulting confidence intervals or distribution of the statistic provide insights about the uncertainty or variability of the sample statistic, without making assumptions about the shape of the population distribution.

Estimate Confidence Intervals: It helps determine the range within which a population parameter (e.g., mean, median, standard deviation) is likely to fall, without relying on strong assumptions about the population distribution.

Assess the Accuracy and Variability of a Statistic: Bootstrapping provides a way to estimate the variability (such as standard errors) of a sample statistic, helping to understand how much it might fluctuate from sample to sample.

Handle Complex Statistics: It is particularly useful when dealing with complex statistics (like medians, regression coefficients, or other non-standard estimators) that may not have easy-to-calculate sampling distributions.

Estimate Bias: Bootstrapping can reveal any potential bias in a sample statistic by comparing the original statistic with the bootstrap distribution.

Perform Hypothesis Testing: In some cases, bootstrapping can be used to generate empirical distributions for testing hypotheses when traditional methods are not feasible or assumptions (such as normality) do not hold.

1. Confidence Intervals and the Null Hypothesis
Null Hypothesis (H 0): In hypothesis testing, the null hypothesis often posits that a population parameter (e.g., mean, difference in means) equals zero. For example, H :=0, where  is the true population mean.

Confidence Interval: A confidence interval gives a range of plausible values for the population parameter based on the sample data. For instance, if you calculate a 95% confidence interval for the mean, it suggests that if you were to repeat the sampling process many times, 95% of the intervals would contain the true population mean.

2. Why Does the Confidence Interval Overlapping Zero Mean "Fail to Reject"?
Interpretation of the Confidence Interval: If the confidence interval includes zero, it indicates that zero is a plausible value for the population parameter given the observed data. In other words, the data do not provide strong enough evidence to rule out zero as the true population mean.

Uncertainty of the Estimate: Even though the sample mean itself is not zero, the confidence interval reflects the uncertainty around that estimate. The sample mean is just one point estimate, and due to sample variability, the true mean could still be zero (or close to it), which is why the confidence interval includes zero.

For example, a sample mean of 2 with a 95% confidence interval of [1.5,5.0] includes zero. This means that, although your sample estimate is 2, you cannot confidently exclude the possibility that the population mean might be zero because zero falls within the interval.
Fail to Reject: If zero is a plausible value for the population mean, there isnt enough evidence to reject the null hypothesis (H 0 :=0). Therefore, you "fail to reject" the null hypothesis.

3. When Would You "Reject the Null Hypothesis"?
Confidence Interval Does Not Include Zero: If the confidence interval does not include zero, it means zero is not a plausible value for the population mean based on the sample data. In this case, the data provide strong evidence that the true population mean is different from zero.

For example, if your sample mean is 2 and the confidence interval is [0.5,4.0], zero is not included in the interval. This indicates that the true population mean is unlikely to be zero, and you would reject the null hypothesis that =0.

This is the opposite situation: the data suggest that the population mean is significantly different from zero, which leads to rejection of the null hypothesis.

4. Why the Sample Mean Being Non-Zero Doesn't Automatically Reject H 
Sampling Variability: The sample mean is just an estimate of the population mean, and its subject to sampling variabilitydifferent samples from the same population can yield different means. The confidence interval accounts for this variability. So, while your sample mean might not be zero, the confidence interval considers the possibility that the population mean might still be zero due to this variability.

Statistical Significance: Statistical significance isn't based solely on the observed sample mean but on how confident you are that the population mean differs from the null hypothesis value. If zero is within the confidence interval, you cant be confident enough to reject the null hypothesis, even if your sample mean is not zero.

5. Connection to Hypothesis Testing and p-Values
Fail to Reject: When the confidence interval includes zero, the p-value from a corresponding hypothesis test would be greater than the significance level (e.g.,=0.05). This means there isnt enough evidence to reject the null hypothesis.

Reject: When the confidence interval does not include zero, the p-value would be less than the significance level (e.g.,=0.05). This provides strong evidence to reject the null hypothesis and conclude that the population mean is likely different from zero.

````python
import numpy as np
import pandas as pd

# Example dataset (InitialHealthScore)
data = df['InitialHealthScore']  # You can change this to any other column or dataset

# Define a function to perform bootstrap
def bootstrap(data, num_bootstrap_samples=1000, confidence_level=0.95, statistic_function=np.mean):
    """
    Function to calculate the bootstrap confidence interval for a given statistic.
    
    Parameters:
    - data: The dataset (array or list of values)
    - num_bootstrap_samples: Number of bootstrap samples to draw
    - confidence_level: The desired confidence level (default is 95%)
    - statistic_function: The function to calculate the statistic of interest (e.g., np.mean, np.median)
    
    Returns:
    - The confidence interval as a tuple (lower_bound, upper_bound)
    """
    
    # Create an array to hold bootstrap samples
    bootstrap_samples = np.random.choice(data, size=(num_bootstrap_samples, len(data)), replace=True)
    
    # Apply the statistic function to each bootstrap sample
    bootstrap_statistics = np.apply_along_axis(statistic_function, 1, bootstrap_samples)
    
    # Calculate percentiles for the confidence interval
    lower_bound = np.percentile(bootstrap_statistics, (1 - confidence_level) / 2 * 100)
    upper_bound = np.percentile(bootstrap_statistics, (1 + confidence_level) / 2 * 100)
    
    return lower_bound, upper_bound

# Calculate the 95% confidence interval for the population mean using bootstrap
mean_ci = bootstrap(data, statistic_function=np.mean)
print(f"95% Confidence Interval for the Mean: {mean_ci}")

# Example: Calculate the 95% confidence interval for the population median
median_ci = bootstrap(data, statistic_function=np.median)
print(f"95% Confidence Interval for the Median: {median_ci}")

# Example: Calculate the 95% confidence interval for the population variance
variance_ci = bootstrap(data, statistic_function=np.var)
print(f"95% Confidence Interval for the Variance: {variance_ci}")
````
