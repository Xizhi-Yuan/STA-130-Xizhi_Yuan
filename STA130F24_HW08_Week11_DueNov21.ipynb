1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

# 1.Accuracy
Best Application Scenario: When the cost of false positives (FP) and false negatives (FN) is roughly the same.
Example:
Weather Forecasting: Predicting if it will rain or not. Incorrect predictions (FP or FN) have minor consequences compared to the overall number of correct predictions.
Rationale: Accuracy gives a general idea of model performance but may be misleading if the data is imbalanced (e.g., far more negatives than positives).

# 2.Sensitivity
Best Application Scenario: When missing a positive case (false negative) is critical.
Example:
Disease Detection: Identifying patients with a specific disease (e.g., cancer). Missing a case (FN) could delay treatment and have severe consequences.
Rationale: High sensitivity ensures most positive cases are captured, even at the expense of more false positives.

# 3.Specificity
Best Application Scenario: When avoiding false positives is more important than missing positives.
Example:
Spam Email Filters: Ensuring legitimate emails (negatives) are not misclassified as spam (FP).
Rationale: High specificity minimizes the likelihood of incorrectly flagging negatives as positives, which could cause inconvenience or harm.

# 4.Precision
Best Application Scenario: When false positives are more critical to avoid than false negatives.
Example:
Fraud Detection in Banking: Identifying fraudulent transactions. Incorrectly labeling legitimate transactions as fraud (FP) could cause significant customer dissatisfaction.
Rationale: High precision ensures that flagged positives are more likely to be truly positive, even if it misses some true positives.

````python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.tree import export_graphviz
import graphviz
import matplotlib.pyplot as plt

# Load and preprocess the dataset
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# Step 1: Drop specified columns
ab = ab.drop(columns=['Weight_oz', 'Width', 'Height'])

# Step 2: Drop rows with NaN entries
ab_reduced_noNaN = ab.dropna()

# Step 3: Convert `Pub year` and `NumPages` to integers
ab_reduced_noNaN['Pub year'] = ab_reduced_noNaN['Pub year'].astype(int)
ab_reduced_noNaN['NumPages'] = ab_reduced_noNaN['NumPages'].astype(int)

# Step 4: Convert `Hard_or_Paper` to category
ab_reduced_noNaN['Hard_or_Paper'] = ab_reduced_noNaN['Hard_or_Paper'].astype('category')

# Define the target variable and features
y = pd.get_dummies(ab_reduced_noNaN["Hard_or_Paper"])['H']  # 1 for Hardcover, 0 for Paperback
X = ab_reduced_noNaN[['NumPages', 'Thick', 'List Price']]

# Train a new DecisionTreeClassifier with max_depth=4
clf2 = DecisionTreeClassifier(max_depth=4, random_state=42)
clf2.fit(X, y)

# Visualize the decision tree using graphviz
dot_data = export_graphviz(
    clf2,
    out_file=None,
    feature_names=['NumPages', 'Thick', 'List Price'],
    class_names=['Paperback', 'Hardcover'],
    filled=True,
    rounded=True,
    special_characters=True
)
graph = graphviz.Source(dot_data)

# Save the visualization to a PDF
graph.render("decision_tree_clf2", format="pdf")  # Saves as decision_tree_clf2.pdf
print("Decision tree saved as 'decision_tree_clf2.pdf'.")

# Inline visualization using Matplotlib
plt.figure(figsize=(20, 10))
plot_tree(
    clf2,
    feature_names=['NumPages', 'Thick', 'List Price'],
    class_names=['Paperback', 'Hardcover'],
    filled=True,
    rounded=True
)
plt.title("Decision Tree Visualization (max_depth=4)")
plt.show()
````

````python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score

# Load and preprocess the dataset
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# Step 1: Drop specified columns
ab = ab.drop(columns=['Weight_oz', 'Width', 'Height'])

# Step 2: Drop rows with NaN entries and explicitly create a copy to avoid warnings
ab_reduced_noNaN = ab.dropna().copy()

# Step 3: Convert `Pub year` and `NumPages` to integers
ab_reduced_noNaN['Pub year'] = ab_reduced_noNaN['Pub year'].astype(int)
ab_reduced_noNaN['NumPages'] = ab_reduced_noNaN['NumPages'].astype(int)

# Step 4: Convert `Hard_or_Paper` to category
ab_reduced_noNaN['Hard_or_Paper'] = ab_reduced_noNaN['Hard_or_Paper'].astype('category')

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(ab_reduced_noNaN, test_size=0.2, random_state=42)

# Train the first model (clf)
y_train_clf = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']
X_train_clf = ab_reduced_noNaN_train[['List Price']]
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X_train_clf, y_train_clf)

# Train the second model (clf2)
y_train_clf2 = y_train_clf  # Same target variable
X_train_clf2 = ab_reduced_noNaN_train[['NumPages', 'Thick', 'List Price']]
clf2 = DecisionTreeClassifier(max_depth=4, random_state=42)
clf2.fit(X_train_clf2, y_train_clf2)

# Prepare the test set features and labels
X_test_clf = ab_reduced_noNaN_test[['List Price']]
y_test_clf = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']

X_test_clf2 = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]
y_test_clf2 = y_test_clf

# Make predictions
y_pred_clf = clf.predict(X_test_clf)
y_pred_clf2 = clf2.predict(X_test_clf2)

# Confusion matrices
cm_clf = confusion_matrix(y_test_clf, y_pred_clf)
cm_clf2 = confusion_matrix(y_test_clf2, y_pred_clf2)

# Compute metrics
sensitivity_clf = recall_score(y_test_clf, y_pred_clf)  # Sensitivity: TP / (TP + FN)
specificity_clf = cm_clf[0, 0] / (cm_clf[0, 0] + cm_clf[0, 1])  # Specificity: TN / (TN + FP)
accuracy_clf = accuracy_score(y_test_clf, y_pred_clf)

sensitivity_clf2 = recall_score(y_test_clf2, y_pred_clf2)
specificity_clf2 = cm_clf2[0, 0] / (cm_clf2[0, 0] + cm_clf2[0, 1])
accuracy_clf2 = accuracy_score(y_test_clf2, y_pred_clf2)

# Display the results
metrics = pd.DataFrame({
    "Model": ["clf (List Price)", "clf2 (NumPages, Thick, List Price)"],
    "Sensitivity": [sensitivity_clf, sensitivity_clf2],
    "Specificity": [specificity_clf, specificity_clf2],
    "Accuracy": [accuracy_clf, accuracy_clf2]
})

print("Confusion Matrix for clf (List Price):")
print(cm_clf)
print("\nConfusion Matrix for clf2 (NumPages, Thick, List Price):")
print(cm_clf2)
print("\nMetrics:")
print(metrics)
````

> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>
> Compared to the simplicity of understanding how different **covariates** contribute towards the final **predicted values** of **multiple linear regression models** (by just reading off the equation to see how predictions work), the the complexity of how all the different **features** interact and combine to together to create the final **predictions** from **classification decision trees** can be staggering. But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. Now we just need to be sure we're not **overfitting** our **classification decision trees** since they can be so **complex**.
````python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# Load and preprocess the dataset
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# Step 1: Drop specified columns
ab = ab.drop(columns=['Weight_oz', 'Width', 'Height'])

# Step 2: Drop rows with NaN entries and explicitly create a copy to avoid warnings
ab_reduced_noNaN = ab.dropna().copy()

# Step 3: Convert `Pub year` and `NumPages` to integers
ab_reduced_noNaN['Pub year'] = ab_reduced_noNaN['Pub year'].astype(int)
ab_reduced_noNaN['NumPages'] = ab_reduced_noNaN['NumPages'].astype(int)

# Step 4: Convert `Hard_or_Paper` to category
ab_reduced_noNaN['Hard_or_Paper'] = ab_reduced_noNaN['Hard_or_Paper'].astype('category')

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(ab_reduced_noNaN, test_size=0.2, random_state=42)

# Define the target and features for clf2
y_train_clf2 = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']
X_train_clf2 = ab_reduced_noNaN_train[['NumPages', 'Thick', 'List Price']]

# Train clf2
clf2 = DecisionTreeClassifier(max_depth=4, random_state=42)
clf2.fit(X_train_clf2, y_train_clf2)

# Extract feature importances from clf2
feature_importances = clf2.feature_importances_

# Use feature names corresponding to feature importances
feature_names = clf2.feature_names_in_

# Create a DataFrame for better visualization
importance_df = pd.DataFrame({
    "Feature": feature_names,
    "Importance": feature_importances
}).sort_values(by="Importance", ascending=False)

# Plot feature importances
plt.figure(figsize=(10, 6))
plt.bar(importance_df["Feature"], importance_df["Importance"])
plt.title("Feature Importances for clf2")
plt.xlabel("Features")
plt.ylabel("Importance")
plt.xticks(rotation=45)
plt.show()

# Display the DataFrame for feature importances
print("Feature Importances for clf2:")
print(importance_df)
````
